```python
# 1a) List the actors (first_name, last_name, actor_id) who acted in more then 25 movies.
cursor.execute("""
    SELECT first_name,last_name,t1.actor_id,COUNT(DISTINCT t1.film_id) as movie_count 
    FROM sakila.film_actor t1 
    JOIN sakila.actor t2 ON t1.actor_id=t2.actor_id 
    GROUP BY first_name,last_name,t1.actor_id 
    HAVING COUNT(DISTINCT film_id)>25
    ORDER BY first_name,last_name;""")
for row in cursor:
    print(row)
```

```python
# 1b) List the actors (first_name, last_name, actor_id) who have worked in German language movies.
cursor.execute("""
    SELECT first_name,last_name,t1.actor_id 
    FROM sakila.film_actor t1 
    JOIN sakila.actor t2 ON t1.actor_id=t2.actor_id 
    JOIN sakila.film t3 ON t1.film_id=t3.film_id 
    JOIN sakila.language t4 ON t3.language_id=t4.language_id 
    WHERE name='German' 
    ORDER BY first_name,last_name;""")
for row in cursor:
    print(row)
```

```python
# 1c) List the actors (first_name, last_name, actor_id) who acted in horror movies and the count of horror movies by each actor.
cursor.execute("""
    SELECT first_name,last_name,t1.actor_id,COUNT(DISTINCT t1.film_id) as horror_movie_count 
    FROM sakila.film_actor t1 
    JOIN sakila.actor t2 ON t1.actor_id=t2.actor_id 
    JOIN sakila.film_category t3 ON t1.film_id=t3.film_id 
    JOIN sakila.category t4 ON t3.category_id=t4.category_id 
    WHERE name='Horror' 
    GROUP BY first_name,last_name,t1.actor_id 
    ORDER BY first_name,last_name;""")
for row in cursor:
    print(row)
```

```python
# 1d) List the customers who rented more than 3 horror movies
cursor.execute("""
    SELECT first_name,last_name,t1.customer_id,COUNT(DISTINCT t3.film_id) as horror_movie_count 
    FROM sakila.customer t1 
    JOIN sakila.rental t2 ON t1.customer_id=t2.customer_id
    JOIN sakila.inventory t3 ON t2.inventory_id=t3.inventory_id 
    JOIN sakila.film_category t4 ON t3.film_id=t4.film_id 
    JOIN sakila.category t5 ON t4.category_id=t5.category_id 
    WHERE name='Horror' 
    GROUP BY first_name,last_name,t1.customer_id 
    HAVING COUNT(DISTINCT t3.film_id)>3
    ORDER BY first_name,last_name;""")
for row in cursor:
    print(row)
```

```python
# 1e) List the customers who rented a movie which starred Scarlett Bening
cursor.execute("""
    SELECT t1.first_name,t1.last_name,t1.customer_id
    FROM sakila.customer t1 
    JOIN sakila.rental t2 ON t1.customer_id=t2.customer_id
    JOIN sakila.inventory t3 ON t2.inventory_id = t3.inventory_id 
    JOIN sakila.film_actor t4 ON t3.film_id = t4.film_id 
    JOIN sakila.actor t5 ON t4.actor_id = t5.actor_id 
    WHERE t5.first_name = 'Scarlett' AND t5.last_name = 'Bening' 
    GROUP BY t1.first_name, t1.last_name, t1.customer_id 
    ORDER BY t1.first_name, t1.last_name;""")
for row in cursor:
    print(row)
```

```python
# 1f) Which customers residing at postal code 62703 rented movies that were Documentaries?
cursor.execute("""
    SELECT t1.first_name, t1.last_name, t1.customer_id
    FROM sakila.customer t1
    JOIN sakila.address t2 ON t1.address_id = t2.address_id
    JOIN sakila.city t3 ON t2.city_id = t3.city_id
    JOIN sakila.rental t4 ON t1.customer_id = t4.customer_id
    JOIN sakila.inventory t5 ON t4.inventory_id = t5.inventory_id
    JOIN sakila.film t6 ON t5.film_id = t6.film_id
    JOIN sakila.film_category t7 ON t6.film_id = t7.film_id
    JOIN sakila.category t8 ON t7.category_id = t8.category_id
    WHERE t2.postal_code = '62703' AND t8.name = 'Documentary'
    ORDER BY t1.first_name, t1.last_name;""")
for row in cursor:
    print(row)
```

```python
# 1g) Find all the addresses (if any) where the second address line is not empty and is not NULL
cursor.execute("""
    SELECT address_id, address2
    FROM sakila.address
    WHERE address2 IS NOT NULL AND address2 != ''
    ORDER BY address2;""")
for row in cursor:
    print(row)
```

```python
# 1h) List the actors (first_name, last_name, actor_id) who played in a film involving a “Crocodile” and a “Shark” (in the same movie)
cursor.execute("""
    SELECT t1.first_name, t1.last_name, t1.actor_id, t3.title, t3.release_year
    FROM sakila.actor t1
    JOIN sakila.film_actor t2 ON t1.actor_id = t2.actor_id
    JOIN sakila.film t3 ON t2.film_id = t3.film_id
    WHERE LOWER(t3.title) LIKE '%crocodile%' AND LOWER(t3.title) LIKE '%shark%'
    ORDER BY t1.first_name, t1.last_name;""")
for row in cursor:
    print(row)
```

```python
# 1i) Find all the film categories in which there are between 55 and 65 films.
cursor.execute("""
    SELECT t1.name AS category_name, COUNT(t3.film_id) AS count_movies
    FROM sakila.category t1
    JOIN sakila.film_category t2 ON t1.category_id = t2.category_id
    JOIN sakila.film t3 ON t2.film_id = t3.film_id
    GROUP BY t1.name
    HAVING COUNT(t3.film_id) BETWEEN 55 AND 65
    ORDER BY COUNT(t3.film_id) DESC, category_name;""")
for row in cursor:
    print(row)
```

```python
# 1j) In which of the film categories is the average difference between the film replacement cost and the rental rate larger than $17?
cursor.execute("""
    SELECT t1.name AS category_name, AVG(t3.replacement_cost - t3.rental_rate) AS mean_diff_replace_rental
    FROM sakila.category t1
    JOIN sakila.film_category t2 ON t1.category_id = t2.category_id
    JOIN sakila.film t3 ON t2.film_id=t3.film_id
    GROUP BY t1.name
    HAVING AVG(t3.replacement_cost - t3.rental_rate) > 17
    ORDER BY AVG(t3.replacement_cost - t3.rental_rate) DESC;""")
for row in cursor:
    print(row)
```

```python
# 1k) Create a list of overdue rentals so that customers can be contacted and asked to return their overdue DVDs.
cursor.execute("""
    SELECT t4.title, t2.first_name, t2.last_name, t5.phone, DATEDIFF(NOW(), t1.rental_date) - t4.rental_duration AS days_overdue
    FROM sakila.rental t1
    JOIN sakila.customer t2 ON t1.customer_id = t2.customer_id
    JOIN sakila.inventory t3 ON t1.inventory_id = t3.inventory_id
    JOIN sakila.film t4 ON t3.film_id = t4.film_id
    JOIN sakila.address t5 ON t2.address_id = t5.address_id
    WHERE t1.return_date IS NULL AND t1.rental_date < DATE_SUB(NOW(), INTERVAL t4.rental_duration DAY)
    ORDER BY t2.first_name, t2.last_name;""")
for row in cursor:
    print(row)
```

```python
# 1l) Find the list of all customers and staff for store_id 1.
cursor.execute("""
    SELECT first_name, last_name, 'staff' AS person_type
    FROM sakila.staff
    WHERE store_id = 1
    UNION
    SELECT first_name, last_name, 'customer' AS person_type
    FROM sakila.customer
    WHERE store_id = 1
    ORDER BY first_name, last_name;""")
for row in cursor:
    print(row)
```

```python
# 2a) List the first and last names of both actors and customers whose first names are the same as the first name of the actor with actor_id 8.
cursor.execute("""
    SELECT first_name, last_name
    FROM sakila.actor
    WHERE first_name = (SELECT first_name FROM sakila.actor WHERE actor_id = 8)
    UNION
    SELECT first_name, last_name
    FROM sakila.customer
    WHERE first_name = (SELECT first_name FROM sakila.actor WHERE actor_id = 8)
    ORDER BY last_name;""")
for row in cursor:
    print(row)
```

```python
# 2b) List customers (first name, last name, customer ID) and payment amounts of customer payments that were greater than average the payment amount.
cursor.execute("""
    SELECT t1.first_name, t1.last_name, t1.customer_id, t2.amount
    FROM sakila.customer t1
    JOIN sakila.payment t2 ON t1.customer_id = t2.customer_id
    WHERE t2.amount > (SELECT AVG(amount) FROM sakila.payment)
    ORDER BY t2.amount DESC;""")
for row in cursor:
    print(row)
```

```python
# 2c) List customers (first name, last name, customer ID) who have rented movies at least once.
cursor.execute("""
    SELECT first_name, last_name, customer_id
    FROM sakila.customer
    WHERE customer_id IN (SELECT DISTINCT customer_id FROM sakila.rental)
    ORDER BY first_name, last_name;""")
for row in cursor:
    print(row)
```

```python
# 2d) Find the floor of the maximum, minimum and average payment amount.
cursor.execute("""
    SELECT 
    FLOOR(MAX(amount)) AS max_payment, 
    FLOOR(MIN(amount)) AS min_payment, 
    FLOOR(AVG(amount)) AS avg_payment 
    FROM sakila.payment;""")
for row in cursor:
    print(row)
```

```python
# 4a) Extract the street number (numbers at the beginning of the address) from the customer address in the customer_list view.
cursor.execute("""
    SELECT address, SUBSTRING_INDEX(address, ' ', 1) AS street_number
    FROM sakila.customer_list
    ORDER BY street_number ASC;""")
for row in cursor:
    print(row)
```

```python
# 4b) List actors (first name, last name, actor id) whose last name starts with characters A, B or C.
cursor.execute("""
    SELECT first_name,last_name,actor_id
    FROM sakila.actor
    WHERE LEFT(last_name, 1) IN ('A', 'B', 'C')
    ORDER BY first_name, last_name ASC;""")
for row in cursor:
    print(row)
```

```python
# 4c) List film titles that contain exactly 10 characters.
cursor.execute("""
    SELECT title
    FROM sakila.film
    WHERE LENGTH(title) = 10
    ORDER BY title ASC;""")
for row in cursor:
    print(row)
```

```python
# 4d) Return a list of distinct payment dates formatted in the date pattern that matches "22/01/2016" (2 digit day, 2 digit month, 4 digit year).
cursor.execute("""
    SELECT DISTINCT DATE_FORMAT(payment_date, '%d/%m/%Y') AS payment_date
    FROM sakila.payment
    ORDER BY payment_date ASC;""")
for row in cursor:
    print(row)
```

```python
# 4e) Find the number of days each rental was out (days between rental_date & return_date), for all returned rentals.
cursor.execute("""
    SELECT rental_id, rental_date, return_date, DATEDIFF(return_date, rental_date) AS days_out
    FROM sakila.rental
    WHERE return_date IS NOT NULL
    ORDER BY days_out DESC;""")
for row in cursor:
    print(row)
```

#### Question 5

```python
# 5a) Which film categories have the highest average rental duration?
cursor.execute("""
    SELECT t1.name AS Category, AVG(t5.return_date - t5.rental_date) AS AverageRentalDuration
    FROM category t1
    JOIN film_category t2 ON t1.category_id = t2.category_id
    JOIN film t3 ON t2.film_id = t3.film_id
    JOIN inventory t4 ON t3.film_id = t4.film_id
    JOIN rental t5 ON t4.inventory_id = t5.inventory_id
    GROUP BY t1.name
    ORDER BY AverageRentalDuration DESC;""")
for row in cursor:
    print(row)
```

```python
# 5b) What is the distribution of customers based on their rental frequency?
cursor.execute("""
    SELECT customer_id, COUNT(rental_id) AS RentalCount
    FROM rental
    GROUP BY customer_id
    ORDER BY RentalCount DESC;""")
for row in cursor:
    print(row)
```

```python
# 5c) What is the revenue generated from each film rating category?
cursor.execute("""
    SELECT t1.rating, SUM(t4.amount) AS TotalRevenue
    FROM film t1
    JOIN inventory t2 ON t1.film_id = t2.film_id
    JOIN rental t3 ON t2.inventory_id = t3.inventory_id
    JOIN payment t4 ON t3.rental_id = t4.rental_id
    GROUP BY t1.rating
    ORDER BY TotalRevenue DESC;""")
for row in cursor:
    print(row)
```

```python
# 5d) Which film categories have the most rentals?
cursor.execute("""
    SELECT t1.name AS Category, COUNT(t5.rental_id) AS RentalCount
    FROM category t1
    JOIN film_category t2 ON t1.category_id = t2.category_id
    JOIN film t3 ON t2.film_id = t3.film_id
    JOIN inventory t4 ON t3.film_id = t4.film_id
    JOIN rental t5 ON t4.inventory_id = t5.inventory_id
    GROUP BY t1.name
    ORDER BY RentalCount DESC;""")
for row in cursor:
    print(row)
```

```python
# 5e) What is the average rental duration for each store?
cursor.execute("""
    SELECT t1.store_id, AVG(t3.return_date - t3.rental_date) AS AverageRentalDuration
    FROM store t1
    JOIN staff t2 ON t1.store_id = t2.store_id
    JOIN rental t3 ON t2.staff_id = t3.staff_id
    GROUP BY t1.store_id;""")
for row in cursor:
    print(row)
```
