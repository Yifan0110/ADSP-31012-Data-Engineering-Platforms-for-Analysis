```python
# 2c) Report the payment_id and payment amount for all payments greater than or equal to $2 and less than or equal to $7, using the BETWEEN operator.
cursor.execute("SELECT payment_id,amount FROM sakila.payment WHERE amount between 2 and 7;")
for row in cursor:
    print(row)
```

```python
# 2d) List all addresses that have phone number that contain digits 589, start with 140, or end with 323. Return all columns in the address table for the matching records.
cursor.execute("SELECT * FROM sakila.address WHERE phone LIKE '%589%' OR phone LIKE '140%' OR phone LIKE '%323';")
for row in cursor:
    print(row)
```

```python
# 2e) List all staff members (first name, last name, email) whose password is NULL
cursor.execute("SELECT first_name,last_name,email FROM sakila.staff WHERE password IS NULL;")
for row in cursor:
    print(row)
```

```python
# 2f) Select all films that have titles containing the word ZOO and have a rental duration greater than or equal to 4. Return the title and rental duration.
cursor.execute("SELECT title,rental_duration FROM sakila.film WHERE title LIKE '%ZOO%' AND rental_duration >= 4;")
for row in cursor:
    print(row)
```

```python
# 2g) What is the cost of renting the movie ACADEMY DINOSAUR for 12 days? Aias the calculated column as rental_cost.
cursor.execute("SELECT title, rental_rate * 12 AS rental_cost FROM sakila.film WHERE title='ACADEMY DINOSAUR';")
result = cursor.fetchone()
print(f"The cost of renting the movie ACADEMY DINOSAUR for 12 days is ${result[1]}")
```

```python
# 2h) List all the unique cities in the address table.
cursor.execute("SELECT DISTINCT city FROM sakila.address a LEFT JOIN sakila.city b ON a.city_id=b.city_id;")
for row in cursor:
    print(row)
```

```python
# 2i) List the first name, last name, and customer ID of the top 10 newest customers across all stores.
cursor.execute("SELECT first_name,last_name,customer_id FROM sakila.customer ORDER BY create_date DESC LIMIT 10;")
for row in cursor:
    print(row)
```

#### Question 3

```python
# 3a) What is the minimum payment received and max payment received across all transactions?
cursor.execute("SELECT min(amount) as min_payment,max(amount) as max_payment FROM sakila.payment;")
result = cursor.fetchone()
print(f"The minumum payment is ${result[0]} and the max payment is ${result[1]} ")
```

```python
# 3b) How many customers rented movies between Feb 2005 & May 2005 (based on rental_date)? Alias the result column as num_customers.
cursor.execute("SELECT COUNT(DISTINCT customer_id) AS num_customers FROM sakila.rental WHERE rental_date BETWEEN '2005-02-01' AND '2005-05-31';")
for row in cursor:
    print(row[0])
```

```python
# 3c)List all movies where replacement_cost is greater than $15 OR the rental_duration is between 6 & 10 days.
cursor.execute("SELECT title, replacement_cost,rental_duration FROM sakila.film WHERE replacement_cost>15 OR rental_duration BETWEEN 6 AND 10;")
for row in cursor:
    print(row)
```

```python
# 3d)What is the total amount spent by customers to rent movies in the year 2005? 
cursor.execute("SELECT SUM(amount) as total_spent FROM sakila.payment WHERE payment_date BETWEEN '2005-01-01' AND '2005-12-31';")
for row in cursor:
    print(row)
```

```python
# 3e)What is the average replacement cost across all movies?
cursor.execute("SELECT AVG(replacement_cost) as avg_replacement_cost FROM sakila.film;")
for row in cursor:
    print(row)
```

```python
# 3f)What is the standard deviation of rental rate across all movies? 
cursor.execute("SELECT STDDEV(rental_rate) AS std_rental_rate FROM sakila.film;")
for row in cursor:
    print(row)
```

```python
# 3g)What is the midrange of the rental duration for all movies?
cursor.execute("SELECT (MIN(rental_duration) + MAX(rental_duration)) / 2 AS midrange_duration FROM sakila.film;")
for row in cursor:
    print(row)
```

#### Question 4

```python
# 4a)List the count of movies that are either G, NC-17, PG-13, PG, or R, grouped by rating.
cursor.execute("SELECT rating, COUNT(DISTINCT film_id) AS movie_count FROM sakila.film WHERE rating IN ('G', 'NC-17', 'PG-13', 'PG', 'R') GROUP BY rating;")
for row in cursor:
    print(row)
```

```python
# 4b)Find the movies where rental rate is greater than $1 and order result set by descending order. 
cursor.execute("SELECT title,rental_rate FROM sakila.film WHERE rental_rate > 1 ORDER BY rental_rate DESC;")
for row in cursor:
    print(row)
```

```python
# 4c)Find the top 2 movies rated R that have the highest replacement costs. 
cursor.execute("SELECT title,rating,replacement_cost FROM sakila.film WHERE rating = 'R'ORDER BY replacement_cost DESC,title LIMIT 2;")
for row in cursor:
    print(row)
```

```python
# 4d) Find the most frequently occurring (mode) rental rate across products.
cursor.execute("SELECT rental_rate, COUNT(rental_rate) FROM sakila.film  GROUP BY rental_rate ORDER BY COUNT(rental_rate) DESC LIMIT 1;")
for row in cursor:
    print(row)
```

```python
# 4e) Find the top 2 longest movies with movie length greater than 50 mins AND which has commentaries as a special features. Return the movie title and length
cursor.execute("SELECT title,length FROM sakila.film WHERE length > 50 AND special_features LIKE '%Commentaries%' ORDER BY length DESC,title LIMIT 2;")
for row in cursor:
    print(row)
```

```python
# 4f) List the top 5 movies (in alphabetical order) whose titles start with Z, are NOT rated R, and have a rental duration less than 7 days.
cursor.execute("SELECT title, rating,rental_duration FROM sakila.film WHERE title LIKE 'Z%' AND rating != 'R' AND rental_duration < 7 ORDER BY title LIMIT 5;")
for row in cursor:
    print(row)
```
